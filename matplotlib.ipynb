## Subplots

```
matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)
```

````python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# instantiates a subplot with 2 rows and 1 column
plt.subplot(2, 1, 1)

# 1o plot
plt.plot(x, y_sin)
plt.title('Sine')

# 2o plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# layout adjustment 
plt.tight_layout()
````

### Layouts

````python
import matplotlib.gridspec as gridspec
````

````python
fig1 = plt.subplots(ncols=2, nrows=2, constrained_layout=True)
````

````python
fig5 = plt.subplots(ncols=3, constrained_layout=True)
````

````python
# Using gridspec and .add_subplot

fig2 = plt.figure(constrained_layout=True)

spec2 = gridspec.GridSpec(ncols=2, nrows=2, figure=fig2)

f2_ax1 = fig2.add_subplot(spec2[0, 0])
f2_ax2 = fig2.add_subplot(spec2[0, 1])
f2_ax3 = fig2.add_subplot(spec2[1, 0])
f2_ax4 = fig2.add_subplot(spec2[1, 1])
````

````python
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(3, 3)

f3_ax1 = fig3.add_subplot(gs[0, :])
f3_ax1.set_title('plt1[0, :]')

f3_ax2 = fig3.add_subplot(gs[1, :-1])
f3_ax2.set_title('plt2[1, :-1]')

f3_ax3 = fig3.add_subplot(gs[1:, -1])
f3_ax3.set_title('plt3[1:, -1]')

f3_ax4 = fig3.add_subplot(gs[-1, 0])
f3_ax4.set_title('plt4[-1, 0]')

f3_ax5 = fig3.add_subplot(gs[-1, -2])
f3_ax5.set_title('plt5[-1, -2]')
````

## Multiple Plots

````python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, color='red', label='Sine')
plt.plot(x, y_cos, color='blue', label='Cosine')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')

# Automatically set the best position for subtitles
plt.legend()
````
